```python
# Importation des bibliothèques nécessaires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Charger les données
data = pd.read_csv('../data/ecg.csv')

# Aperçu des données
print("Aperçu des données :")
display(data.head())

# Dimensions et informations sur le dataset
print(f"Nombre d'échantillons : {data.shape[0]}, Nombre de colonnes : {data.shape[1]}")
print("\nInformations sur les colonnes :")
display(data.info())

# Statistiques descriptives
print("\nStatistiques descriptives :")
display(data.describe())

# Vérification des valeurs manquantes
missing_values = data.isnull().sum().sum()
print(f"\nNombre de valeurs manquantes : {missing_values}")

# Distribution des classes
print("\nDistribution des classes (dernière colonne) :")
class_distribution = data.iloc[:, -1].value_counts()
print(class_distribution)

# Visualisation de la distribution des classes
plt.figure(figsize=(6, 4))
sns.countplot(x=data.iloc[:, -1], palette="viridis")
plt.title("Distribution des classes (Normal vs Anormal)")
plt.xlabel("Classe")
plt.ylabel("Nombre d'échantillons")
plt.show()

# Analyse de la variance des caractéristiques
variances = data.var(axis=0).sort_values(ascending=False)[:10]
print("\nTop 10 des caractéristiques avec la variance la plus élevée :")
print(variances)

# Visualisation des caractéristiques avec la variance la plus élevée
plt.figure(figsize=(8, 6))
variances.plot(kind='bar', color='orange')
plt.title("Top 10 des caractéristiques avec la plus grande variance")
plt.ylabel("Variance")
plt.xlabel("Caractéristiques")
plt.show()

# Matrice de corrélation (zoom sur les corrélations fortes)
plt.figure(figsize=(10, 8))
corr_matrix = data.iloc[:, :-1].corr()
sns.heatmap(corr_matrix, cmap='coolwarm', cbar=True, annot=False)
plt.title("Matrice de corrélation des caractéristiques")
plt.show()

# Relations entre caractéristiques fortement corrélées
high_corr = corr_matrix[corr_matrix > 0.8].stack().reset_index()
high_corr = high_corr[high_corr['level_0'] != high_corr['level_1']]
print("\nPaires de caractéristiques fortement corrélées (> 0.8) :")
display(high_corr)

# Distribution de plusieurs caractéristiques
sample_cols = data.columns[:5]
plt.figure(figsize=(12, 8))
for i, col in enumerate(sample_cols):
    plt.subplot(2, 3, i+1)
    sns.histplot(data[col], kde=True, bins=30, color=np.random.rand(3,))
    plt.title(f"Distribution de {col}")
plt.tight_layout()
plt.show()

# Analyse avec PCA (Réduction de dimension)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data.iloc[:, :-1])

# Appliquer PCA
pca = PCA(n_components=2)
data_pca = pca.fit_transform(scaled_data)

# Ajouter les labels pour la visualisation
pca_df = pd.DataFrame(data_pca, columns=['PCA1', 'PCA2'])
pca_df['Class'] = data.iloc[:, -1]

# Visualisation PCA
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Class', data=pca_df, palette="Set1", alpha=0.7)
plt.title("Projection des données après PCA")
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.legend(title="Classe", loc="best")
plt.show()

# Analyse de l'influence des classes sur une caractéristique
plt.figure(figsize=(8, 6))
sns.boxplot(x=data.iloc[:, -1], y=data.iloc[:, 0], palette="viridis")
plt.title("Boxplot de la première caractéristique par classe")
plt.xlabel("Classe")
plt.ylabel("Valeurs de la première caractéristique")
plt.show()

# Distribution des classes sur plusieurs dimensions (PCA en 3D)
from mpl_toolkits.mplot3d import Axes3D

pca_3d = PCA(n_components=3)
data_pca_3d = pca_3d.fit_transform(scaled_data)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(data_pca_3d[:, 0], data_pca_3d[:, 1], data_pca_3d[:, 2],
                     c=data.iloc[:, -1], cmap="viridis", alpha=0.8)
legend = ax.legend(*scatter.legend_elements(), title="Classes")
ax.add_artist(legend)
plt.title("Projection 3D des données (PCA)")
plt.show()

# Variance expliquée par PCA
explained_variance = pca.explained_variance_ratio_ * 100
print("\nPourcentage de variance expliquée par chaque composante PCA :")
print(f"Composante 1 : {explained_variance[0]:.2f}%, Composante 2 : {explained_variance[1]:.2f}%")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.animation as animation

# Charger les données
data = pd.read_csv('../data/ecg.csv')

# Standardiser les données (exclure la colonne des classes)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data.iloc[:, :-1])

# Réduction de dimension avec PCA
pca = PCA(n_components=2)
data_pca = pca.fit_transform(scaled_data)

# Ajouter les colonnes PCA et les classes au DataFrame
pca_df = pd.DataFrame(data_pca, columns=['PCA1', 'PCA2'])
pca_df['Class'] = data.iloc[:, -1]  # Classes réelles

# Simuler une notion de "temps"
pca_df['Time'] = np.linspace(0, 100, len(pca_df))  # Temps simulé

# Initialiser l’animation
fig, ax = plt.subplots(figsize=(8, 6))
sns.set(style="whitegrid")

# Limites des axes
xlim = (pca_df['PCA1'].min() - 1, pca_df['PCA1'].max() + 1)
ylim = (pca_df['PCA2'].min() - 1, pca_df['PCA2'].max() + 1)

# Créer un graphique vide pour l'animation
scat = ax.scatter([], [], c=[], cmap="coolwarm", alpha=0.7)
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.set_title("Évolution des états des patients dans l'espace PCA")
ax.set_xlabel("PCA1")
ax.set_ylabel("PCA2")

# Fonction de mise à jour pour chaque frame
def update(frame):
    current_data = pca_df[pca_df['Time'] <= frame]  # Filtrer les points jusqu'à l'instant simulé
    scat.set_offsets(current_data[['PCA1', 'PCA2']].values)  # Mettre à jour les positions
    scat.set_array(current_data['Class'].values)  # Mettre à jour les couleurs
    return scat,

# Animation
ani = animation.FuncAnimation(fig, update, frames=np.linspace(0, 100, 200), blit=True)

# Sauvegarder ou afficher l'animation
ani.save('pca_evolution.gif', writer='imagemagick')  # Sauvegarder en GIF
plt.show()
```
